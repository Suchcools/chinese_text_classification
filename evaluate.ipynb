```python
import pandas as pd
import numpy as np
data_list=[]
data_label=['close-set','open-set']
data_list.append(np.load('/home/linjw/code/Z_Text_Classification/roc_save2/0.6973.npz'))
data_list.append(np.load('/home/linjw/code/Z_Text_Classification/roc_save2/0.7719.npz'))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
plt.rc('font',family='Times New Roman')
plt.figure(figsize=(8, 6)) # 设置画布大小

# 循环绘制每个ROC曲线
for i in range(2):
    y_pred = data_list[i]['y1']
    y_true = data_list[i]['y2']
    fpr, tpr, thresholds = roc_curve(y_true, y_pred)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=2, label='%s (AUC = %0.4f)' % (data_label[1-i], roc_auc))

# 绘制对角线
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

# 设置坐标轴范围和标签
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate',size=15)
plt.ylabel('True Positive Rate',size=15)
plt.title('ROC Curve of Two Task',size=20)
plt.legend(loc="lower right",fontsize=15)
plt.show()
```
